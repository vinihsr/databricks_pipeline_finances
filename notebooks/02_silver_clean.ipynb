```python
from pyspark.sql import functions as F
from pyspark.sql.types import DecimalType
```

```python
spark_df = spark.read.table("workspace.bronze.bronze_stocks_prices")

spark_df = spark_df.withColumn("Date", F.col("Date").cast("date"))
spark_df = spark_df.withColumn("Close", F.col("Close").cast(DecimalType(10, 2)))
spark_df = spark_df.withColumn("High", F.col("High").cast(DecimalType(10, 2)))
spark_df = spark_df.withColumn("Low", F.col("Low").cast(DecimalType(10, 2)))
spark_df = spark_df.withColumn("Open", F.col("Open").cast(DecimalType(10, 2)))

spark_df = spark_df.dropDuplicates(subset=["Date", "Ticker"])
spark_df = spark_df.filter(F.col("Date").isNotNull())

spark_df.write.format("delta").mode("overwrite").saveAsTable("workspace.silver.silver_stocks_prices")

spark.sql("SELECT Date, Ticker, Close, High, Low, Open, Volume FROM workspace.default.silver_stocks_prices").show()
```
